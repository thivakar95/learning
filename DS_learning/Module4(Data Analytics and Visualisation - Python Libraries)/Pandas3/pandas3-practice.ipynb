```python
director_df = pd.read_csv("directors.csv",index_col=0)
movies_df = pd.read_csv("movies.csv",index_col=0)
```

```python
final_df= movies_df.merge(director_df,left_on='director_id',right_on='id').drop(['id_x','id_y'],axis=1)
```

```python
final_df['director_name'].unique(),final_df['director_name'].nunique()
```

```python
#Apply function

def encode(gender):
    if gender == 'Male':
        return 'm'
    else:
        return 'f'
```

```python
final_df['gender'].apply(encode).unique()
```

```python
final_df['gender_map'] = final_df['gender'].apply(encode)
```

```python
final_df
```

```python
def prof(x):
    return x['revenue']-x['budget']

final_df['profit'] = final_df.apply(prof,axis=1)
final_df
```
