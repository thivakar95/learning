```python
data_df = pd.melt(pfizer_df,id_vars=['Date','Drug_Name','Parameter'],var_name='time',value_name='reading')
```

```python
data_tigdy=data_df.pivot(index=['Date','time','Drug_Name'],columns='Parameter',values='reading').reset_index()
```

```python
data_tigdy
```

```python
def temp_mean(x):
    x['Temperature_avg'] = x['Temperature'].mean()
    return x
```

```python
data_tigdy.groupby('Drug_Name')['Temperature'].mean()
```

```python
data_tigdy = data_tigdy.groupby(['Drug_Name']).apply(temp_mean)
data_tigdy
```

```python
data_tigdy=data_tigdy.drop('Drug_Name',axis=1).reset_index()
```

```python
data_tigdy[data_tigdy['Temperature'].isnull()]
```

```python
data_tigdy['Temperature'].fillna(data_tigdy['Temperature_avg'],inplace=True)
```

```python
data_tigdy.isnull().sum()
```

```python
def pressure_avg(x):
    x['Pressure_avg'] = x['Pressure'].mean()
    return x
```
