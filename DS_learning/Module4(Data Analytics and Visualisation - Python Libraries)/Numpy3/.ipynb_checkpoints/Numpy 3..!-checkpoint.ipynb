```python
import numpy as np

arr = np.array([[1, 2, 3], [4, 5, 6]])
contiguous_arr = np.ascontiguousarray(arr)

print(f"Original Array Memory Address: {hex(id(arr))}")
print(f"Contiguous Array Memory Address: {hex(id(contiguous_arr))}")
```

```python
import numpy as np

# Create a NumPy array
arr = np.array([[1, 2, 3], [4, 5, 6]])

# Iterate through the elements and print their memory addresses
for row in arr:
    for elem in row:
        print(f"Element {elem}: Memory Address {hex(id(elem))}")
```

```python
arr = np.arange(6).reshape(2, 3)

print(arr.data.contiguous)  # True
print(arr.data.c_contiguous)  # True
print(arr.data.f_contiguous)  # False
```
