# Join

```python
users = pd.DataFrame({"userid":[1, 2, 3], "name":["sharadh", "shahid", "khusalli"]})
users
```

```python
msgs = pd.DataFrame({"userid":[1, 1, 2, 4], "msg":['hmm', "acha", "theek hai", "nice"]})
msgs
```

```python
pd.concat([users,msgs])
```

```python
pd.concat([users,msgs],axis=1)
```

```python
msgs
```

```python
users
```

```python
users.merge(msgs,on='userid')# inner join
```

```python
users.merge(msgs,on='userid',how='outer')# outer join
```

```python
users.merge(msgs,on='userid',how='left')# left join
```

```python
users.merge(msgs,on='userid',how='right')# right join
```

```python
users
```

```python
msgs
```

```python
users.rename(columns = {"userid":"id"}, inplace = True)
users
```

```python
users.merge(msgs,left_on='id',right_on='userid')# inner join
```

# How many Directors are making the movies actually

```python
directors = pd.read_csv('directors.csv', index_col = 0)
directors.head()
```

```python
directors.shape
```

```python
movies['director_id'].value_counts()
```

```python
movies['director_id'].nunique()
```

```python
movies['director_id'].isin(directors['id'])
```

```python
np.all(movies['director_id'].isin(directors['id']))
```

```python
data = movies.merge(directors,how='left',left_on='director_id',right_on='id')
```

```python
data.head(3)
```

```python
data.drop(['director_id','id_y'],axis=1,inplace=True)
```
