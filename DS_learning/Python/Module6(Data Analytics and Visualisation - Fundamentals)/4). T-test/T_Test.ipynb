```python
import pandas as pd

df_iq = pd.read_csv('/content/iq_two_schools.csv')
df_iq.head()
```

```python
df_iq.groupby('School')['iq'].mean()
```

```python
df_iq.groupby('School')['iq'].count()
```

```python
iq_1=df_iq[df_iq['School']=='school_1']['iq']
iq_2=df_iq[df_iq['School']=='school_2']['iq']
```

```python
from scipy.stats import ttest_ind
```

```python
ttest_ind(iq_1,iq_2)
```

```python
ttest_ind(iq_1,iq_2,alternative="greater")
```

```python
ttest_ind(iq_1,iq_2,alternative="less")
```

```python
ttest_ind(iq_2,iq_1,alternative="less")
```
