```python
import pandas as pd
import numpy as np
data=pd.read_csv('final_vg.csv')
data.head()
```

```python
# count of each genre

cat_counts=data['Genre'].value_counts()
```

```python
x_bar=cat_counts.index
y_bar=cat_counts

#plt.bar(x_bar[:5],y_bar[:5]) #top 5
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_bar,y_bar)
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_bar,y_bar,color="Olive",width=0.2)
plt.xticks(rotation=90)
plt.xlabel("Genre")
plt.ylabel("Count",fontsize=24)
plt.title("distribution of videogames by genre")
plt.show()
```

```python
plt.bar(x_bar,y_bar,width=0.2)
```

```python
sns.countplot(x="Genre",data=data,order=data['Genre'].value_counts().sort_values(ascending=True).index,color='cornflowerblue')
plt.xticks(rotation=90)
plt.ylabel("this is y value")
plt.show()
plt.savefig("temp.png")
```
