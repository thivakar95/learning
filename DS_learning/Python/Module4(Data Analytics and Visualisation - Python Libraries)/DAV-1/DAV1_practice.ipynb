```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/home/thivakaranb/Pictures/DS_learning/DAV-1/Archive/final_vg.csv")
```

```python
data.head()
```

```python
x_bar = data['Genre'].value_counts().index
y_bar = data['Genre'].value_counts()
x_bar
```

```python
plt.figure(figsize=(15,8))
plt.bar(x_bar,y_bar,color='g',width=.8)
plt.title("Top 15 category of Genre",fontsize=20)
plt.xticks(rotation=90)
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.show()
```

```python
plt.figure(figsize=(15,8))
sns.countplot(x='Genre',data=data,order=data['Genre'].value_counts().index)
plt.xticks(rotation=90)
plt.xlabel("Genre",fontsize=20)
plt.ylabel("Count",fontsize=20)
plt.title("Top category of Genre",fontsize=30)
# plt.xticklabels()
plt.savefig("category_of_genre.png")
plt.show()
```
