```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("Archive2/vgsales.csv")
```

```python
data.head()
```

```python
data.info()
```

```python
data['Name'].value_counts()
```

```python
ih=data.loc[data['Name']=='Ice Hockey']
```

```python
sns.lineplot(x='Year',y='NA_Sales',data=ih,color="orange")
plt.title("Ice Hockey NA Sales",fontsize=20)
plt.ylabel("NA Sales",fontsize=15)
plt.xlabel("Year",fontsize=15)
plt.xlim(left=1990,right=2010) # to set limitled data for the graph
plt.show()
```

```python
baseball=data.loc[data['Name']=='Baseball']
```

plt.figure(figsize=(15,8))
plt.title("Sales data for Ice Hockey and Baseball",fontsize=20)
plt.ylabel("NA sales", fontsize=12)
plt.xlabel("year",fontsize=12)
sns.lineplot(x='Year',y='NA_Sales',data=baseball,label='Baseball',color='red')
sns.lineplot(x='Year',y='NA_Sales',data=ih,label='Ice Hockey',color='blue')
# plt.legend(["Ice Hockey", "Baseball"]) #this option is not beautiful
plt.legend(loc="upper left")
plt.text(x=2002,y=max(ih['NA_Sales']),s="Peak point")
plt.grid()

```python
plt.figure(figsize=(15,8))
plt.title("Sales data for Ice Hockey and Baseball",fontsize=20)
plt.ylabel("NA sales", fontsize=12)
plt.xlabel("year",fontsize=12)
sns.lineplot(x='Year',y='NA_Sales',data=baseball,label='Baseball',color='red')
sns.lineplot(x='Year',y='NA_Sales',data=ih,label='Ice Hockey',color='blue')
# plt.legend(["Ice Hockey", "Baseball"]) #this option is not beautiful
plt.legend(loc="upper left")
plt.text(x=2002,y=max(ih['NA_Sales']),s="Peak point")
plt.grid()
```
