```python
top3_pub=data['Publisher'].value_counts().index[:3]
top3_gen=data['Genre'].value_counts().index[:3]
top3_plat=data['Platform'].value_counts().index[:3]
top3_data=data.loc[(( data["Publisher"].isin(top3_pub) & data["Genre"].isin(top3_gen) )
                   & data["Platform"].isin(top3_plat))
                  ]  
top3_data
```

```python
## HOw can we compare the top3 platforms these publishers use?
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Publisher',hue='Platform' ,data=top3_data)
```

```python
plt.figure(figsize=(12,8))
sns.countplot(x='Publisher',hue='Platform' ,data=top3_data,dodge=False)
```
