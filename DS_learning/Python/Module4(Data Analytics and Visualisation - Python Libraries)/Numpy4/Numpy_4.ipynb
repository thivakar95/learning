```python
import numpy as np
def add_padding(mat):

   res = None
   #no. of rows
   n = len(mat)

   #no. of columns
   m = len(mat[0])

   #adding a layer of 0s to the lower boundary
   intmdt_mat1 = np.vstack((mat,[0]*m))

   #adding a layer of 0s to the upper boundary
   intmdt_mat2 = np.vstack(([0]*m,intmdt_mat1))

   #array of shape (n+2,1)
   a = np.array([0]*(n+2)).reshape(n+2,1)

   #adding a layer of 0s on the left boundary
   intmdt_mat3 = np.hstack((intmdt_mat2, a))

   #adding a layer of 0s on the right boundary
   res = np.hstack((a, intmdt_mat3))
   return res
```
